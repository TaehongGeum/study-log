# Categoricals

astype()은 데이터의 형을 바꿔주는 메소드이다.  
astype(category)는 데이터의 형태를 object에서 category로 바꾼다는 뜻이다.  
category를 사용하는 이유는  
1. 메모리 절약  
2. 의미 부여(순서 만들기) 가능  


```python
df = pd.DataFrame(
    {'id': [1, 2, 3, 4, 5, 6], 'raw_grade': ['a', 'b', 'b', 'a', 'a', 'e']}
)
df['grade'] = df['raw_grade'].astype('category')
df['grade']
```

.cat은 카테고리 데이터만을 위한 전용 버튼이다.  
.cat이 없으면 뒤에 나오는 .rename_categories같은 명령어를 컴퓨터가 알아 듣지 못한다.  
.rename_categories()는 기존에 설정된 카테고리의 이름을 새로운 이름으로 싹 바꾸는 역할을 한다.

.set_categories()는 카테고리의 범위와 순서를 강제로 다시 정의하는 명령이다.  
역할  
1. 빈 버켓 미리 만들어 두기  
2. 내 마음대로 순서(서열)정하기.

```python
new_categories = ['very good', 'good', 'very bad']
df['grade'] = df['grade'].cat.rename_categories(new_categories)
df['grade'] = df['grade'].cat.set_categories(
    ['very bad', 'bad', 'medium', 'good', 'very good']
)
df['grade']
```

```python
df.sort_values(by='grade')
```

observed= 이 코드는 눈에 보이는 것만 믿을 건지 아니면 우리가 정한 명단을 믿을 건지 정하는 코드이다.  
observed=True는 데이터에 실제로 존재하는 값만 집계한다.  
observed=False는 데이터에 실제로 관측되지 않아도 카테고리 리스트에 있다면 무조건 표시한다.  
.size는 데이터의 내용(값)이 있든 없든 상관없이, 그 그룹에 속한 행(Raw)이 몇 줄인지를 세는 것이다.


```python
df.groupby('grade', observed=False).size()
```

# Plot

plt.close('all')는 지금까지 열려있는 모든 그래프 창(Figure)을 강제로 닫고, 메모리에서 삭제하라는 명령어이다.  
.cumsum()은 누적 합을 구하는 코드이다. 쉽게 말해 지금까지 나온 값을 계속 더해 쌓아가라는 의미이다.  

ts.plot(); 이 코드는 판다스에게 데이터(ts)를 가지고 선 그래프를 그려달라는 명령어이다.  
이 코드 안에는 3가지 자동화 기능이 있다.  
1. X축 자동 배정: 인덱스를 자동으로 X축 값으로 잡는다.  
2. Y축 자동 배정: 데이터의 값을 Y축으로 잡는다.  
3. 선 연결: 점으로 찍기만 하는게 아니라 점을 선으로 이어준다.  
  
;의 역할  
;이 없으면 복잡한 텍스트가 같이 출력된다.  
;이 있으면 텍스트를 숨기고 깔끔한 그래프 그림만 보여준다.

```python
import matplotlib.pyplot as plt
plt.close('all')
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot();
```

plt.figure();은 이전 그림은 잊어버리고 새롭고 깨끗한 창(Canvas)를 하나 띄워라라는 의미이다.  
plt.legend(loc='best');는 범례를 가장 명당자리에 알아서 배치하라는 명령어이다.  
여기서 loc는 위치를 의미한다.  
loc='best'는 자동 추천 기능이다.

```python
df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=['A', 'B', 'C', 'D']
)
df = df.cumsum()
plt.figure();
df.plot();
plt.legend(loc='best');
```
